# **My first ml project**


## load data set




```python
import pandas as pd
df=pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/refs/heads/master/delaney_solubility_with_descriptors.csv')
df
```

## **data preparation**

### data seperation x and y

```python
y=df['logS']
print(y)
```

```python
x=df.drop('logS',axis=1)
x
```

### data spliiting


```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=100)
```

```python
x_train
```

```python
x_test
```

## **model training**

### linear regression

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

**train model**

```python
y_lr_train_pred=lr.predict(x_train)
y_lr_test_pred=lr.predict(x_test)
```

```python
y_lr_train_pred
```

```python
y_lr_test_pred
```

**evaluate model performancee**

```python
from sklearn.metrics import mean_squared_error,r2_score
lr_train_mse=mean_squared_error(y_train,y_lr_train_pred)
lr_train_r2=r2_score(y_train,y_lr_train_pred)
lr_test_mse=mean_squared_error(y_test,y_lr_test_pred)
lr_test_r2=r2_score(y_test,y_lr_test_pred)
```

```python
print('lr train mse:',lr_train_mse)
print('lr train r2:',lr_train_r2  )
print('lr test mse:',lr_test_mse)
print('lr test r2:',lr_test_r2)
```

```python
lr_result=pd.DataFrame(['linear regression',lr_train_mse,lr_train_r2,lr_test_mse,lr_test_r2]).transpose()
lr_result.columns=['method','train MSE','train r2','test MSE','test R2']
lr_result
```

# **render forest**

# **Model training**

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor(max_depth=2,random_state=100)
rf.fit(x_train,y_train)
```

### apllying the model to make a predection

```python
y_rf_train_pred=rf.predict(x_train)
y_rf_test_pred=rf.predict(x_test)
```

```python
y_rf_train_pred
```

```python
y_rf_test_pred
```

```python
from sklearn.metrics import mean_squared_error,r2_score
rf_train_mse=mean_squared_error(y_train,y_rf_train_pred)
rf_train_r2=r2_score(y_train,y_rf_train_pred)
rf_test_mse=mean_squared_error(y_test,y_rf_test_pred)
rf_test_r2=r2_score(y_test,y_rf_test_pred)
```

```python
rf_result=pd.DataFrame(['linear regression',rf_train_mse,rf_train_r2,rf_test_mse,rf_test_r2]).transpose()
rf_result.columns=['method','train MSE','train r2','test MSE','test R2']
rf_result
```

```python
df_models=pd.concat([lr_result,rf_result],axis=0).reset_index()
df_models.drop('index',axis=1,inplace=True)
df_models
```
